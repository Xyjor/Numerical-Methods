```python
def newton_raphson_sqrt(N, tolerance=0.003):
    # Initial guess
    x = N / 2.0 if N > 1 else 1.0
    iterations = []
    while True:
        iterations.append(x)
        next_x = 0.5 * (x + N / x)
        if abs(next_x - x) < tolerance:
            iterations.append(next_x)
            break
        x = next_x
    return iterations

numbers = [3, 7, 20, 99]
for num in numbers:
    iters = newton_raphson_sqrt(num)
    print(f"Square root of {num}:")
    for i, val in enumerate(iters):
        print(f"  Iteration {i}: {val:.6f}")
    print()
    
```

```python
def newton_raphson_sqrt_custom(N, x0, tolerance=0.003):
    x = x0
    iterations = []
    while True:
        iterations.append(x)
        next_x = 0.5 * (x + N / x)
        if abs(next_x - x) < tolerance:
            iterations.append(next_x)
            break
        x = next_x
    return iterations

guesses = [(3, 1.5), (7, 2.5), (20, 4.5), (99, 10.0)]
for num, guess in guesses:
    iters = newton_raphson_sqrt_custom(num, guess)
    print(f"Square root of {num}:")
    for i, val in enumerate(iters):
        print(f"  x_{i} = {val:.5f}")
    print()
```

```python
import tkinter as tk
from tkinter import messagebox
import numpy as np
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk

def calculate_root():
    result_text.delete(1.0, tk.END)
    
    try:
        N = float(entry_number.get())
        n = int(entry_degree.get()) 
        x0 = float(entry_guess.get())
        tolerance = float(entry_tolerance.get())
        
        # Basic validation
        if n < 2:
            messagebox.showerror("Invalid Input", "Root degree must be 2 or greater.")
            return
        if N < 0 and n % 2 == 0:
            messagebox.showerror("Invalid Input", f"Cannot calculate the real {n}th root of a negative number.")
            return
        if x0 == 0:
            messagebox.showerror("Invalid Input", "Initial guess cannot be zero.")
            return

        x = x0
        iteration = 1
        
        # Lists to store points for plotting
        x_points = [x0]
        y_points = [x0**n - N]
        
        result_text.insert(tk.END, f"Finding {n}th root of {N}...\n")
        result_text.insert(tk.END, f"Initial guess (x0): {x0}\n")
        result_text.insert(tk.END, f"Tolerance: {tolerance}\n")
        result_text.insert(tk.END, "-"*35 + "\n")
        
        while True:
            # 1. Calculate f(x) and f'(x) explicitly for the report
            f_x = (x ** n) - N
            f_prime_x = n * (x ** (n - 1))
            
            # 2. Prevent division by zero if the derivative hits exactly 0
            if f_prime_x == 0:
                result_text.insert(tk.END, "\nError: Derivative became zero. Method failed.")
                break
                
            # 3. Newton-Raphson step: x_new = x_old - f(x)/f'(x)
            next_x = x - (f_x / f_prime_x)
            
            # 4. Calculate the Error (Absolute Difference)
            error_diff = abs(next_x - x)
            
            x_points.append(next_x)
            y_points.append(next_x**n - N)
            
            # Display detailed tabular output for the lab report
            result_text.insert(tk.END, f"Iteration {iteration}:\n")
            result_text.insert(tk.END, f"  x_k      = {x:.5f}\n")
            result_text.insert(tk.END, f"  f(x_k)   = {f_x:.5f}\n")
            result_text.insert(tk.END, f"  f'(x_k)  = {f_prime_x:.5f}\n")
            result_text.insert(tk.END, f"  x_k+1    = {next_x:.5f}\n")
            result_text.insert(tk.END, f"  Error    = {error_diff:.5f}\n\n")
            
            # Check stopping condition
            if error_diff < tolerance:
                result_text.insert(tk.END, "-"*35 + "\n")
                result_text.insert(tk.END, f"Final Answer: ≈ {next_x:.5f}\n")
                result_text.insert(tk.END, f"Tolerance reached at Iteration {iteration}.")
                break
                
            x = next_x
            iteration += 1
            
            # Safety break
            if iteration > 50:
                result_text.insert(tk.END, "\nStopped at 50 iterations.")
                break
        
        # --- Update the Graph ---
        ax.clear()
        
        max_val = max(abs(x0), abs(next_x)) * 1.5
        min_val = min(0, x0)
        if min_val == 0:
            min_val = -1 
            
        x_vals = np.linspace(min_val, max_val, 400)
        y_vals = x_vals**n - N
        
        ax.plot(x_vals, y_vals, label=f'f(x) = x^{n} - {N}', color='blue')
        ax.axhline(0, color='black', linewidth=1)
        
        # Plot steps
        for i in range(len(x_points) - 1):
            curr_x = x_points[i]
            curr_y = y_points[i]
            next_x_pt = x_points[i+1]
            
            ax.plot([curr_x, curr_x], [0, curr_y], 'k--', alpha=0.5)
            ax.plot(curr_x, curr_y, 'ro')
            ax.plot([curr_x, next_x_pt], [curr_y, 0], 'r-', alpha=0.7)
        
        final_root = x_points[-1]
        ax.plot(final_root, 0, 'go', markersize=8, label=f'Root ≈ {final_root:.3f}')
        
        ax.set_title(f"Visualizing {n}th Root of {N}")
        ax.set_xlabel("x")
        ax.set_ylabel("f(x)")
        ax.legend()
        ax.grid(True, linestyle=':', alpha=0.7)
        
        canvas.draw()

    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numeric values.")

# --- GUI Setup ---
root = tk.Tk()
root.title("Newton-Raphson Calculator (Lab Report Edition)")
root.geometry("950x650") 

left_frame = tk.Frame(root, width=320)
left_frame.pack(side=tk.LEFT, fill=tk.Y, padx=15, pady=15)

right_frame = tk.Frame(root)
right_frame.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)

input_frame = tk.Frame(left_frame)
input_frame.pack(fill=tk.X)

tk.Label(input_frame, text="Number (N):").grid(row=0, column=0, sticky="w", pady=5)
entry_number = tk.Entry(input_frame, width=15)
entry_number.grid(row=0, column=1, pady=5)
entry_number.insert(0, "5") 

tk.Label(input_frame, text="Root Degree (n):").grid(row=1, column=0, sticky="w", pady=5)
entry_degree = tk.Entry(input_frame, width=15)
entry_degree.grid(row=1, column=1, pady=5)
entry_degree.insert(0, "3") 

tk.Label(input_frame, text="Initial Guess (x0):").grid(row=2, column=0, sticky="w", pady=5)
entry_guess = tk.Entry(input_frame, width=15)
entry_guess.grid(row=2, column=1, pady=5)
entry_guess.insert(0, "2.0")

tk.Label(input_frame, text="Tolerance:").grid(row=3, column=0, sticky="w", pady=5)
entry_tolerance = tk.Entry(input_frame, width=15)
entry_tolerance.grid(row=3, column=1, pady=5)
entry_tolerance.insert(0, "0.003")

calc_button = tk.Button(left_frame, text="Calculate & Plot", command=calculate_root, bg="#4CAF50", fg="white", font=("Arial", 10, "bold"))
calc_button.pack(pady=15, fill=tk.X)

scrollbar = tk.Scrollbar(left_frame)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
result_text = tk.Text(left_frame, height=25, width=40, yscrollcommand=scrollbar.set, font=("Courier", 10))
result_text.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
scrollbar.config(command=result_text.yview)

# --- Right Frame: Matplotlib Graph ---
fig = Figure(figsize=(5, 5), dpi=100)
ax = fig.add_subplot(111)

canvas = FigureCanvasTkAgg(fig, master=right_frame)
canvas_widget = canvas.get_tk_widget()
canvas_widget.pack(side=tk.TOP, fill=tk.BOTH, expand=True)

toolbar = NavigationToolbar2Tk(canvas, right_frame)
toolbar.update()
canvas_widget.pack(side=tk.TOP, fill=tk.BOTH, expand=True)

ax.set_title("Newton-Raphson Method Visualization")
ax.grid(True, linestyle=':', alpha=0.7)

root.mainloop()
```
