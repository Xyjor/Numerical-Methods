```python
import numpy as np

# 1. Set variables
h = 0.01
x_val = 1.0  # Change this to your target x value

# 2. Define the functions
def f1(x):
    return (x**2 + 1) / np.sin(x)

def f2(x):
    return np.sin(x) / (x**3 - 3*x**5)

def f3(x):
    numerator = np.cos(x) * np.sin(x)
    denominator = 3*x**2 + (np.cos(x + 2) / x**2)
    return numerator / denominator

functions = [
    (f1, "F(x) = (x^2 + 1) / sin(x)"),
    (f2, "F(x) = sin(x) / (x^3 - 3x^5)"),
    (f3, "F(x) = (cos(x)sin(x)) / (3x^2 + cos(x+2)/x^2)")
]

# 3. Calculate and print step-by-step solutions
for f, name in functions:
    fx = f(x_val)
    fx_plus_h = f(x_val + h)
    fx_minus_h = f(x_val - h)
    
    fdd = (fx_plus_h - fx) / h
    bdd = (fx - fx_minus_h) / h
    cdd = (fx_plus_h - fx_minus_h) / (2 * h)
    
    print(f"--- Solving for {name} ---")
    print(f"F({x_val})      = {fx:.5f}")
    print(f"F({x_val} + {h}) = {fx_plus_h:.5f}")
    print(f"F({x_val} - {h}) = {fx_minus_h:.5f}\n")
    
    print(f"FDD = ({fx_plus_h:.5f} - {fx:.5f}) / {h} = {fdd:.5f}")
    print(f"BDD = ({fx:.5f} - {fx_minus_h:.5f}) / {h} = {bdd:.5f}")
    print(f"CDD = ({fx_plus_h:.5f} - {fx_minus_h:.5f}) / {2*h} = {cdd:.5f}")
    print("=" * 55 + "\n")
```

# Numerical Differentiation Solutions
**Given Parameters:**
* Evaluation Point ($x$): 1.0
* Step Size ($\Delta x$): 0.01

---

## 1. $F(x) = \frac{x^2 + 1}{\sin(x)}$

### Step 1: Evaluate Function at Specific Points
* $F(1.00) = \frac{1^2 + 1}{\sin(1)} \approx 2.376841$
* $F(1.01) = \frac{1.01^2 + 1}{\sin(1.01)} \approx 2.385481$
* $F(0.99) = \frac{0.99^2 + 1}{\sin(0.99)} \approx 2.368473$

### Step 2: Calculate Derivatives
* **Forward Difference (FDD):**
  $$\frac{F(1.01) - F(1.00)}{0.01} = \frac{2.385481 - 2.376841}{0.01} = 0.8640$$
* **Backward Difference (BDD):**
  $$\frac{F(1.00) - F(0.99)}{0.01} = \frac{2.376841 - 2.368473}{0.01} = 0.8368$$
* **Central Difference (CDD):**
  $$\frac{F(1.01) - F(0.99)}{0.02} = \frac{2.385481 - 2.368473}{0.02} = 0.8504$$
